### Creating Merged dataset

```python
result_driver = pd.merge(results_df, drivers_df, how='left',on='driverId', suffixes=('_res','_drv'))


result_driver = pd.merge(result_driver, status_df, how='left', on='statusId', suffixes=('_resDrv', '_st'))

result_driver_race = pd.merge(result_driver,races_df,how='left',on='raceId', suffixes=('_resDrv','_rac'))


result_driver_race_circuit = pd.merge(result_driver_race, circuits_df,how='left', on='circuitId', suffixes=('_resDrvRce','_cir'))

merged_df = pd.merge(result_driver_race_circuit, constructors_df,how='left',on='constructorId', suffixes=('_resDrvRceCir','_con'))
```

```python
pd.set_option('display.max_columns', None)
merged_df.head()
```

```python
merged_df.columns
```

#### Delete unncessary columns:

```python
merged_df = merged_df.drop(columns=['number_res','laps','fastestLap','rank','round',\
                                    'statusId','number_drv','url_resDrv','url_rac',\
                                    'fp1_date', 'fp1_time','fp2_date','fp2_time','fp3_date', 'fp3_time',\
                                    'quali_date','quali_time', 'sprint_date', 'sprint_time',\
                                    'url_resDrvRceCir','url_con'])
```

```python
pd.set_option('display.max_columns', None)
merged_df.head()
```

#### Columns Cleaning:

- If position == "\\N", it means that the driver didn't end the race (collision, engine, disqualification...)
- Valid positions are from 1 to 22, we replace "\\N" by 0

```python
merged_df['position'] = merged_df['position'].replace({'\\N':0})
merged_df['position'] = pd.to_numeric(merged_df['position'])
```

- In status, any status of ending with +x laps, means that the driver ended the race (fishied) but with a delay of x laps from the winner of the race => can change all these status by Finished

```python
merged_df['status'] = merged_df.status.str.replace("\+\d\s\w*|\+\d\d\s\w*", 'Finished')
```

#### Renaming columns

```python
merged_df.columns
```

```python
merged_df.rename(columns={"code": "driverCode", "forename": "driverFirstName","surname": "driverLastName",
                            "dob": "driverDateBirth", "nationality_resDrvRceCir": "driverNationality",
                            "name_resDrvRce":"raceName", "date": "raceDate","time_rac": "timeRace",
                            "name_cir": "circuitName", "name": "constructorName", "nationality_con": "constructtorNationality"
                            }, inplace=True)
```
