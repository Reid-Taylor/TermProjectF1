```julia
using CSV, DataFrames, Plots, StatsBase, Gurobi, JuMP, LinearAlgebra
```

```julia
data = select(CSV.read("F1_final_dataset.csv", DataFrame), Not("Column1"))
```

```julia
laps = CSV.read("./Dataset/lap_times.csv", DataFrame);
```

```julia
group = groupby(laps,[:raceId, :driverId])
mappings = DataFrame(raceId = [], driverId = [], milliseconds = [])
for obj in group
    push!(mappings, [obj[1, :raceId], obj[1, :driverId], sum(obj[!,:milliseconds])])
end
```

```julia
data[!,:milliseconds] = zeros(size(data)[1])
for i in 1:size(data)[1]
    for j in 52:168
        if data[i,j] == 1
            race = mappings[mappings[:,:raceId] .== data[i,:raceId],:]
            time = race[race[:,:driverId] .== j-50,:]
            mill = time[!,:milliseconds]
            if length(mill) â‰  0
                data[i,:milliseconds] = mill[1]
            end
        end
    end
end
```

```julia
CSV.write("final_with_time.csv", data)
```
